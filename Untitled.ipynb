```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Lenovo\Downloads\stack-overflow-developer-survey-2025\survey_results_public25.csv")

print(df.shape)
df.head(3)
```

```python
useful_cols = [
    "MainBranch", "EdLevel", "Employment", "WorkExp", "YearsCode",
    "LanguageHaveWorkedWith", "LanguageWantToWorkWith",
    "DatabaseHaveWorkedWith", "DatabaseWantToWorkWith",
    "PlatformHaveWorkedWith", "PlatformWantToWorkWith",
    "WebframeHaveWorkedWith", "WebframeWantToWorkWith",
    "DevEnvsHaveWorkedWith", "DevEnvsWantToWorkWith",
    "AIModelsHaveWorkedWith", "AIModelsWantToWorkWith",
    "ConvertedCompYearly"
]

# filter dataframe
df_filtered = df[useful_cols]

# save cleaned dataset (optional)
df_filtered.to_csv(r"C:\Users\Lenovo\Downloads\survey_filtered.csv", index=False)

# preview
print(df_filtered.shape)
df_filtered.head()
```

```python
df_filtered.isnull().sum()
```

```python
# Define skill columns
skill_cols = [
    "LanguageHaveWorkedWith", "LanguageWantToWorkWith",
    "DatabaseHaveWorkedWith", "DatabaseWantToWorkWith",
    "PlatformHaveWorkedWith", "PlatformWantToWorkWith",
    "WebframeHaveWorkedWith", "WebframeWantToWorkWith",
    "DevEnvsHaveWorkedWith", "DevEnvsWantToWorkWith",
]

# 1. Drop rows where ALL skill columns are missing
df_filtered = df_filtered.dropna(subset=skill_cols, how="all")

# 2. Fill missing categorical columns with "Unknown"
df_filtered["EdLevel"] = df_filtered["EdLevel"].fillna("Unknown")
df_filtered["Employment"] = df_filtered["Employment"].fillna("Unknown")
df_filtered["WorkExp"] = df_filtered["WorkExp"].fillna("Unknown")
df_filtered["YearsCode"] = df_filtered["YearsCode"].fillna("Unknown")

# 3. Fill missing skills with empty string
df_filtered[skill_cols] = df_filtered[skill_cols].fillna("")

# 4. Convert semicolon-separated strings into list of skills
def split_skills(cell):
    if cell == "":
        return []
    return [s.strip() for s in cell.split(";")]

for col in skill_cols:
    df_filtered[col] = df_filtered[col].apply(split_skills)

# 5. Handle salary (optional: drop or fill with 0)
df_filtered["ConvertedCompYearly"] = df_filtered["ConvertedCompYearly"].fillna(0)

# Save cleaned dataset
df_filtered.to_csv(r"C:\Users\Lenovo\Downloads\survey_cleaned.csv", index=False)

print(df_filtered.shape)
df_filtered.head(3)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

skill_columns = [
    "LanguageHaveWorkedWith", "LanguageWantToWorkWith",
    "DatabaseHaveWorkedWith", "DatabaseWantToWorkWith",
    "PlatformHaveWorkedWith", "PlatformWantToWorkWith",
    "WebframeHaveWorkedWith", "WebframeWantToWorkWith",
    "DevEnvsHaveWorkedWith", "DevEnvsWantToWorkWith",
]
encoded_dfs = []

# loop through each skill col and one-hot encode
for col in skill_columns:
    mlb = MultiLabelBinarizer()
    df_encoded = pd.DataFrame(mlb.fit_transform(df_filtered[col].dropna()),
                              columns=[f"{col}_{skill}" for skill in mlb.classes_],
                              index=df_filtered[col].dropna().index)
    encoded_dfs.append(df_encoded)

# combine all encoded features into one big dataframe
df_skills = pd.concat(encoded_dfs, axis=1)

# merge with salary (optional)
final_df = pd.concat([df_skills, df_filtered["ConvertedCompYearly"]], axis=1)

print(final_df.shape)
final_df.head()
```

```python
import pandas as pd
import joblib

# --------------------------
# Load your processed dataset
# --------------------------
# final_df = pd.read_csv("final_df.csv")

# Skill columns (exclude salary col)
skill_cols = [col for col in final_df.columns if col != "ConvertedCompYearly"]

# --------------------------
# 1. Static seeds (domain knowledge)
# --------------------------
role_seed_skills = {
    "Data Scientist": ["Python", "SQL", "R"],
    "ML Engineer": ["Python", "C++", "TensorFlow"],
    "Backend Developer": ["Java", "C#", "Node.js"],
    "Cloud Engineer": ["Amazon Web Services (AWS)", "Microsoft Azure", "Google Cloud"],
    "Cybersecurity Engineer": ["Python", "Linux", "Wireshark"]
}

# --------------------------
# 2. Map skill names to dataset cols
# --------------------------
def generate_skill_mappings(columns):
    skill_to_col, col_to_skill = {}, {}
    for col in columns:
        if col == "ConvertedCompYearly":
            continue
        if "_" in col:
            skill_name = col.split("_", 1)[-1]
        else:
            skill_name = col
        skill_to_col[skill_name] = col
        col_to_skill[col] = skill_name
    return skill_to_col, col_to_skill

skill_to_col, col_to_skill = generate_skill_mappings(skill_cols)

# --------------------------
# 3. Dynamic expansion (find related skills)
# --------------------------
role_to_skills = {}

for role, seed_skills in role_seed_skills.items():
    expanded = []
    for seed in seed_skills:
        seed_col = skill_to_col.get(seed)  # dataset column for seed
        if seed_col is None:
            continue  

        # Find correlations
        corr = final_df[skill_cols].corrwith(final_df[seed_col]).sort_values(ascending=False)

        # Pick top 5 related skills (excluding itself)
        related = corr.drop(seed_col).head(5).index.tolist()
        expanded.extend(related)

    # --- Always work in dataset col space ---
    all_skills = [skill_to_col.get(s, s) for s in seed_skills if s in skill_to_col]
    all_skills.extend(expanded)

    # --- Deduplicate while preserving order ---
    seen = set()
    full_skills = []
    for s in all_skills:
        if s not in seen:
            full_skills.append(s)
            seen.add(s)

    role_to_skills[role] = full_skills

```

```python
import pandas as pd
import joblib

# --------------------------
# Load processed dataset
# --------------------------
# final_df = pd.read_csv("final_df.csv")
skill_cols = [col for col in final_df.columns if col != "ConvertedCompYearly"]

# --------------------------
# 1. Skill Name Normalization Map
# --------------------------
skill_name_map = {
    "Amazon Web Services (AWS)": "AWS",
    "Microsoft Azure": "Azure",
    "Google Cloud Platform (GCP)": "Google Cloud",
    "Bash/Shell": "Bash/Shell (all shells)",
    "Jupyter": "Jupyter Notebook/JupyterLab"
}

def normalize_skill(skill):
    return skill_name_map.get(skill, skill)

# --------------------------
# 2. Domain-specific whitelists
# --------------------------
role_whitelists = {
    "Data Scientist": [
        "Python", "SQL", "R", "Pandas", "NumPy", "TensorFlow", 
        "PyTorch", "Scikit-learn", "Jupyter Notebook/JupyterLab", 
        "Matplotlib", "Seaborn"
    ],
    "ML Engineer": [
        "Python", "C++", "TensorFlow", "PyTorch", "Scikit-learn", 
        "Keras", "Pandas", "NumPy", "MLflow", "Docker", "Kubernetes"
    ],
    "Backend Developer": [
        "Java", "C#", "Node.js", "Spring", "Express", "SQL", 
        "PostgreSQL", "MongoDB", "REST", "GraphQL", "Docker"
    ],
    "Cloud Engineer": [
        "AWS", "Azure", "Google Cloud", "Kubernetes", "Docker", 
        "Terraform", "Linux", "Python", "Bash/Shell (all shells)"
    ],
    "Cybersecurity Engineer": [
        "Python", "Linux", "Wireshark", "Metasploit", "Burp Suite", 
        "Kali Linux", "Bash/Shell (all shells)", "Firewalls", "SIEM"
    ]
}

# --------------------------
# 3. Generate skill mappings
# --------------------------
def generate_skill_mappings(columns):
    skill_to_col, col_to_skill = {}, {}
    for col in columns:
        if col == "ConvertedCompYearly":
            continue
        if "_" in col:
            skill_name = col.split("_", 1)[-1]
        else:
            skill_name = col
        skill_name = normalize_skill(skill_name)  # normalize
        skill_to_col[skill_name] = col
        col_to_skill[col] = skill_name
    return skill_to_col, col_to_skill

skill_to_col, col_to_skill = generate_skill_mappings(skill_cols)

# --------------------------
# 4. Hybrid: whitelist filter + correlation expansion
# --------------------------
role_to_skills = {}

for role, allowed_skills in role_whitelists.items():
    expanded = []
    for seed in allowed_skills:
        seed_col = skill_to_col.get(seed)
        if seed_col is None:
            continue
        # correlations
        corr = final_df[skill_cols].corrwith(final_df[seed_col]).sort_values(ascending=False)
        related = corr.drop(seed_col).head(5).index.tolist()
        expanded.extend(related)

    # keep only whitelist skills
    expanded_clean = [col_to_skill.get(s, s) for s in expanded if col_to_skill.get(s, s) in allowed_skills]

    # deduplicate, preserving order
    full_skills = list(dict.fromkeys(allowed_skills + expanded_clean))

    role_to_skills[role] = full_skills

# --------------------------
# 5. Compute absolute average salaries per skill
# --------------------------
skill_salary = {}
for skill, col in skill_to_col.items():
    if col in final_df.columns:
        avg_salary = final_df.loc[final_df[col] == 1, "ConvertedCompYearly"].mean()
        if pd.notna(avg_salary):
            skill_salary[skill] = round(avg_salary, 0)

# --------------------------
# 6. Save mappings
# --------------------------
joblib.dump(role_to_skills, "role_to_skills_hybrid.pkl")
joblib.dump(skill_to_col, "skill_to_col.pkl")
joblib.dump(col_to_skill, "col_to_skill.pkl")
joblib.dump(skill_salary, "skill_salary.pkl")

print("✅ Hybrid + Whitelist + Salary mapping created!")
for role, skills in role_to_skills.items():
    print(f"\n{role}: {skills}")

print("\n💰 Sample salaries:")
for k, v in list(skill_salary.items())[:10]:
    print(f"{k}: ${v}")
```
